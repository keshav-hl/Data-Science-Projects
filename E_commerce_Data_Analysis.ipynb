```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("/content/Superstore data.csv.csv", encoding='latin1')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

Type Conversion of the columns

---



```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day'] = data['Order Date'].dt.day
data['Order day of week'] = data['Order Date'].dt.dayofweek
```

```python
data.head()
```

:1. You need to calculate the monthly sales of the store and identify which month had the highest sales and which month had the lowest sales.

---



```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
sales_by_month
```

```python
sales_by_month.max()
```

```python
fig = px.line(sales_by_month,
              x='Order Month',
              y='Sales',
              title='Monthly Sales')
fig.show()
```

2. You need to analyze sales based on product categories and determine
which category has the lowest sales and which category has the highest sales.


---



```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
fig = px.pie(sales_by_category,
             values = 'Sales',
             names = 'Category',
             hole = 0.2,
             color_discrete_sequence = px.colors.qualitative.Pastel)
fig.update_traces(textposition = 'inside', textinfo = 'percent+label')
fig.update_layout(title = 'Sales by Category', title_font = dict(size=24))
fig.show()
```

3. The sales analysis needs to be done based on sub-categories.

---



```python
data.head()
```

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
sales_by_subcategory
```

```python
fig = px.bar(sales_by_subcategory, x = 'Sub-Category', y = 'Sales', title = 'Sales by Sub-Category')
fig.show()
```

4. You need to analyze the monthly profit from sales and determine which month had the highest profit.


---



```python
sales_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()
```

```python
sales_by_month
```

```python
fig = px.line(sales_by_month, x = 'Order Month', y = 'Profit', title = 'Profit by Month')
fig.show()
```

5. Analyze the profit by category and sub-category.

---



```python
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
```

```python
profit_by_category
```

```python
fig = px.pie(profit_by_category, values = 'Profit', names = 'Category', hole = 0.2, color_discrete_sequence = px.colors.qualitative.Pastel)
fig.update_traces(textposition = 'inside', textinfo = 'percent+label')
fig.update_layout(title = 'Profit by Category', title_font = dict(size=24))
fig.show()
```

```python
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()
```

```python
profit_by_subcategory
```

```python
fig = px.bar(profit_by_subcategory, x = 'Sub-Category', y = 'Profit', title = 'Profit by Sub-Category')
fig.show()
```

6. Analyze the sales and profit by customer segment.

---



```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment
```

```python
color_palette = colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(
    x=sales_profit_by_segment['Segment'],
    y=sales_profit_by_segment['Sales'],
    name='Sales',
    marker_color=color_palette[0]
))

fig.add_trace(go.Bar(
    x=sales_profit_by_segment['Segment'],
    y=sales_profit_by_segment['Profit'],
    name='Profit',
    marker_color=color_palette[1]
))

fig.update_layout(
    title='Sales and Profit Analysis by Customer Segment',
    xaxis_title='Customer Segment',
    yaxis_title='Amount'
)

fig.show()
```

7. Analyze the sales to profit ratio.

---



```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['Sales_to_Profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```

```python
fig = px.bar(sales_profit_by_segment, x = 'Segment', y = 'Sales_to_Profit_Ratio', title = 'Sales to Profit Ratio by Segment')
fig.show()
```
